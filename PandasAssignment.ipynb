```python
import pandas as pd
```

```python
# Load dataset
data = pd.read_csv('vgsales.csv')
# Display the first 10 rows
data.head(10)
```

```python
# Display the shape
print(data.shape)
```

```python
# Display the basic information about the dataset
data.info()
```

```python
# Check for missing data 
missing_values = data.isnull()
print(missing_values.sum())
```

```python
# Replace missing values (N/A) with blank
data_blank = data.fillna(" ")
#print(data_nona)
```

```python
# Filter rows with blank in the year column

select_year_blank = data_blank["Year"] == " "
filter_year_blank = data[select_year_blank]
filter_year_blank.head()
```

```python
filter_year_blank.shape
```

```python
# Filter rows with blank in the publisher column

select_publish_blank = data_blank["Publisher"] == " "

filter_publish_blank = data[select_publish_blank]
filter_publish_blank.head()
```

```python
filter_publish_blank.shape
```

```python
# Display the top 5 video games by global sale
data_blank_sorted = data_blank.sort_values(by = "Global_Sales", ascending = False)
data_blank_sorted.head(5)
```

```python
# Calculate the total sales in North America for the 'Action' genre
# NA_Sales (column) - Sales in North America (in millions)

# Filter rows where Genre == Action

action_video_games = (data_blank["Genre"] == "Action")

filter_action_vg = data_blank[action_video_games]

filter_action_vg.head(12)
```

```python
# Total NA_Sales

# select NA_sales column from the filtered (genre == action) data
na_action_vg = filter_action_vg["NA_Sales"]

#na_action_vg.shape

total_NA_sales = na_action_vg.sum()
print(f'The total North American sales on Action video games is: {total_NA_sales}')
```

```python
# Find the top 3 publishers by global sales

data_sort_gs = data_blank.sort_values(by = "Global_Sales", ascending = False)
data_sort_gs.head(3)
```

```python
# Obtain the unique values of the genre column

data_blank["Genre"].unique()
```

```python
# Calculate the average sales in Europe for the 'Shooter' genre

#1. Filter rows for which genre == shooter

shooter_video_games = (data_blank["Genre"] == "Shooter")
data_blank[shooter_video_games]
```

```python
#2. Select the Europe column from the filtered data

shooter_videogames_df = data_blank[shooter_video_games]

europe_sales = shooter_videogames_df["EU_Sales"]

#3. Calculate the average

average_eu_sales = europe_sales.mean()

print(f'The average European sales for the shooter video games is {average_eu_sales}')
```

```python
# Find the game with the highest sales in Japan in the 'Sports' genre.

#1. Filter rows for which genre == sports

data_sports = data_blank["Genre"] == "Sports"
data_sports_df = data_blank[data_sports]

sports_jpsales = data_sports_df["JP_Sales"]

sports_jpsales.max()
```
