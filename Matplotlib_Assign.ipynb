```python
# Convert Revenue series from string to float

sales["Revenue"] = sales["Revenue"].str.replace(',', '').astype(float)

          # OR

#sales = sales.Revenue.str.replace(',', '').astype(float)
```

```python
# Summarize numerical variables

sales.describe()

# All the three numerical variables have their mean = median

# There are therefore no suspicious outliers
```

```python
# Box plot 

plt.figure(figsize = (8, 5))
plt.subplots_adjust(wspace = 0.3)
plt.subplot(1,3,1)
plt.boxplot(sales["Revenue"])
plt.title("Revenue Boxplot")

plt.subplot(1,3,2)
plt.boxplot(sales["Orders"])
plt.title("Orders Boxplot")

plt.subplot(1,3,3)
plt.boxplot(sales["Average_Order_Value"])
plt.title("Average Order Value Boxplot")

plt.show()

# From the box-plots there are no outliers in the three numerical variables
```

```python
# Create a line plot showing the monthly revenue over the two-year period.
  # No style

plt.plot(sales["Month"], sales["Revenue"], color = "#adad3b", linewidth = 1, marker = ".")
plt.xlabel("Time")
plt.ylabel("Monthly revenue")
plt.title("Monthly revenue from January 2020 to December 2021")
plt.xticks(rotation = 90)
#plt.grid(True)
plt.show()
```

```python
# Create a bar plot showing the total number of orders per month

plt.bar(sales["Month"], sales["Orders"], color = "#adad3b")
plt.xlabel("Time")
plt.ylabel("Total number of orders per month")
plt.title("Total monthly orders")
plt.xticks(rotation = 90)
#plt.grid(True)
plt.show()
```

```python
# Create a scatter plot showing the relationship between the number of orders and 
  # the average order value.
    
plt.scatter(sales["Orders"], sales["Average_Order_Value"], color = "#adad3b")
plt.xlabel("Total number of orders per month")
plt.ylabel(" Average order value")
plt.title("Scatter plot of average order value by total monthly orders")
#plt.grid(True)
plt.show()    
```

```python
# Create a subplot containing all three visualizations created in steps 2-4

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize = (7, 20))
plt.subplots_adjust(hspace = 0.5)

# Line chart
ax1.plot(sales["Month"], sales["Revenue"], color = "#adad3b", linewidth = 1, marker = ".")
ax1.set_xlabel("Time")
ax1.set_ylabel("Monthly revenue")
ax1.set_title("Monthly revenue from January 2020 to December 2021")
ax1.tick_params(rotation = 90)

# Bar chart
ax2.bar(sales["Month"], sales["Orders"], color = "#adad3b")
ax2.set_xlabel("Time")
ax2.set_ylabel("Total number of orders per month")
ax2.set_title("Total monthly orders")
ax2.tick_params(rotation = 90)

# scatter plot

ax3.scatter(sales["Orders"], sales["Average_Order_Value"], color = "#adad3b")
ax3.set_xlabel("Total number of orders per month")
ax3.set_ylabel(" Average order value")
ax3.set_title("Scatter plot of average order value by total monthly orders")

#Save the subplot as a high-resolution image file.
plt.savefig('monthly_sales.png', dpi=325, format = "jpeg")
plt.show()
```
